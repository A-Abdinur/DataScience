```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def draw_regress(x):
    prediction = intercept + x*slope
    return prediction

x = np.random.normal(0,1,100)
y = 100 - (x+np.random.normal(0,0.1,100))*3 #variation added by adding a random number
plt.scatter(x,y)

slope, intercept, r_value, p_value, std_err= stats.linregress(x,y)
fitline = draw_regress(x) #assigning y-values using regression calcs
plt.plot(x,fitline, color = "red")
print(r_value**2) #
plt.show()
```
