```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from scipy import stats
from sklearn.metrics import r2_score
np.random.seed(1)#####
x = np.random.normal(3,1,100)
y= (np.random.normal(50,10,100)/x) 

x = np.array(x)
y=np.array(y)
p4 = np.poly1d(np.polyfit(x,y,6)) ########
xp = np.linspace(0, 6, 100) #??????

plt.scatter(x,y)
plt.plot(xp,p4(xp)) #plotted the fitted model to get prediction using a range of x to be predicted
plt.show()

r2 = r2_score(y, p4(x)) #calculating fit using y values from model and data ??? why p4(x) and not p4(xp)
r2
```
